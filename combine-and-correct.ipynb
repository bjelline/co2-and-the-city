```python
%matplotlib inline
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
from parse import *
```

```python
map_df1 = pd.read_csv("biggest-msa-to-county-mapping.csv")
map_df1.columns
```

```python
map_df2 = map_df1
map_df2.columns = [u'MSA according to Glaeser and Kahn', u'MSA/CMSA Code', u'PMSA Code', u'FIPS State/County Code', u'Census MCD Code', u'FIPS MCD Code', u'Name of Area', u'Population', u'Population1990', u'ChangeNumber', u'ChangePercent']
map_df2
```

```python
map_df3 = map_df2
del(map_df3['Population1990'])
del(map_df3['ChangeNumber'])
del(map_df3['ChangePercent'])
map_df3['Population'] = map_df3['Population'].replace(",","",regex=True).astype(int)

map_df3.dtypes
```

```python
msa_of_county = {}
msa_of_county_fips = {}
msa_fips_of_county_fips = {}
msa_data = []
county_data = []

row = {}    
msa = None
msa_code = None

i=0

for i, row in map_df3.iterrows():
    if type(row['MSA according to Glaeser and Kahn']) == str:
        #print "Found a MSA %s" % row['Selected MSA'] 
        msa_data.append(dict(row))
        msa = row['MSA according to Glaeser and Kahn']
        msa_code = row['MSA/CMSA Code']
        #print msa_code,msa
    elif row['MSA/CMSA Code'] == msa_code:
        #print row['MSA/ CMSA Code'], row['FIPS State/ County Code'], row['Area']
        msa_of_county[row['Name of Area']] = msa
        
        #print row['FIPS State/ County Code']
        if ~np.isnan(row['FIPS State/County Code']):
            msa_of_county_fips[int(row['FIPS State/County Code'])] = msa
            msa_fips_of_county_fips[int(row['FIPS State/County Code'])] = int(msa_code)
        county_data.append(dict(row))
    lastrow = row
print "found %d MSAs and %d counties in all" % (len(msa_data), len(county_data))
```

```python
msa_df = pd.DataFrame(msa_data)
del(msa_df['FIPS State/County Code'])
del(msa_df['PMSA Code'])
del(msa_df['Census MCD Code'])
del(msa_df['FIPS MCD Code'])
msa_df.rename(columns={'Population':'Population in MSA'}, inplace=True)

print "created dataframe with %d MSAs" % len(msa_df)
msa_df.dtypes
```

```python
msa_df.head()
```

```python
county_df1 = pd.DataFrame(county_data)
del(county_df1['PMSA Code'])
del(county_df1['MSA according to Glaeser and Kahn'])
county_df1.rename(columns={'Population':'Population in County (part)'}, inplace=True)

print "created dataframe with %d counties" % len(county_df1)
county_df1.dtypes
```

```python
county_df1[county_df1['FIPS State/County Code']==9005]
```

```python
county_df2 = county_df1[ ~np.isnan(county_df1["FIPS State/County Code"]) ]
county_df3 = county_df2[ np.isnan(county_df2["FIPS MCD Code"]) ]

county_df3["FIPS State/County Code"] = county_df3["FIPS State/County Code"].astype('int')
print "Selected dataframe down to %d counties" % len(county_df3)
county_df3[county_df3['FIPS State/County Code']==9005]
```

```python
county_df4 = county_df3
del(county_df4["Census MCD Code"])
del(county_df4["FIPS MCD Code"])
county_df4[county_df4['FIPS State/County Code']==9005]
```

```python
census_df = pd.read_csv("R10843310_SL050.csv", usecols=["Geo_QName","Geo_FIPS","SE_T003_002","SE_T003_001","SE_T003_003","SE_T020_001"])
census_df.rename(columns={'Geo_QName':'County, State', 
                           'Geo_FIPS': 'FIPS', 
                           'SE_T003_002': "Census Population", 
                           'SE_T003_001': "Population Density", 
                           'SE_T003_003': "Area (Land)", 
                           'SE_T020_001': "Households"}, inplace=True)

census_df.head()
```

```python
county_and_census_df1 = pd.merge(county_df4, census_df,how="left",left_on="FIPS State/County Code", right_on="FIPS")
county_and_census_df1.head()
```

## split up counties

These Counties occure more than once. they are split up into separate MASs

```python
parts = county_and_census_df1[county_and_census_df1["Name of Area"].str.contains("part")]
print "%d counties are not containt in whole, but just in parts of MSA" % len(parts)
parts
```

```python
county_and_census_df2 = county_and_census_df1
county_and_census_df2["county part"] = county_and_census_df2["Population in County (part)"] /  county_and_census_df2["Census Population"] 
county_and_census_df2[county_and_census_df2["MSA/CMSA Code"]==3280]
```

```python
county_and_census_df2[county_and_census_df2['FIPS State/County Code']==9005]
```

```python
county_and_census_df = county_and_census_df2
del(county_and_census_df["FIPS"])
county_and_census_df["Area (Land)"] = county_and_census_df2["Area (Land)"] * county_and_census_df2["county part"]
county_and_census_df["Households"] = (county_and_census_df2["Households"] * county_and_census_df2["county part"]).astype("int")
county_and_census_df[county_and_census_df["MSA/CMSA Code"]==3280]
```

```python
emissions_df = pd.read_csv("vulcan.csv", usecols=['State', ' County', ' FIPS', ' Total', 'Commercial', 'Industrial', 'Residential', 'Electricity Prod', 'Onroad', 'Cement', 'Aircraft', 'Airborne', 'Nonroad', 'Pop 2000'])
emissions_df.rename(columns={' County'  :'County', 
                       ' FIPS'    :'FIPS', 
                       ' Total'   :'Total', 
                       'Pop 2000' : "Population whole County According to Vulcan"}, inplace=True)
emissions_df.columns
```

## Merge County and Emission
This needs to be a left join, as some counties are split up in serveral parts
for now we will copy the emission data to every part of the county.
For Las Vegas this works out ok: the three counties that make up Las Vegas 
are all 100% located in Las Vegas.

```python
county_and_emissions_df1 = pd.merge(county_and_census_df, emissions_df,how="inner",left_on="FIPS State/County Code", right_on="FIPS")
county_and_emissions_df1[county_and_emissions_df1["MSA/CMSA Code"]==4120]
```

But now an example where this does not work: In MSA 3280 = Hartford, CT we find only parts
of Windham and Hartford and Litchfield

```python
county_and_emissions_df1[county_and_emissions_df1["MSA/CMSA Code"]==3280][["FIPS State/County Code","MSA/CMSA Code","Name of Area","Population in County (part)","Census Population","Population whole County According to Vulcan","county part","Total"]]
```

### Correct the emission data
If only 20% of Litchfield County is inside this MSA, 
then we only want to count 20% of the emissions here.

```python
county_and_emissions_df = county_and_emissions_df1
for col in ['Total', 'Commercial', 'Industrial', 'Residential', 'Electricity Prod', 'Onroad', 'Cement', 'Aircraft', 'Airborne', 'Nonroad']:
    county_and_emissions_df["%s Emission" % col] = county_and_emissions_df1[col] * county_and_emissions_df1["county part"]
county_and_emissions_df[county_and_emissions_df["MSA/CMSA Code"]==3280][["FIPS State/County Code","MSA/CMSA Code","Name of Area","Population in County (part)","Census Population","Population whole County According to Vulcan","county part","Total Emission"]]   
```

### Add MSA Name and MSA FIPS
to the county data

```python
county_and_emissions_df2 = county_and_emissions_df
county_and_emissions_df2["MSA FIPS"] = county_and_emissions_df2.FIPS.replace(msa_fips_of_county_fips)
county_and_emissions_df2["MSA"]      = county_and_emissions_df2.FIPS.replace(msa_of_county_fips)
county_and_emissions_df2[county_and_emissions_df2["MSA/CMSA Code"]==4120][["Name of Area", "FIPS","MSA", "MSA FIPS", "Population in County (part)", "Total"]]
```

```python
county_and_emissions_df2.columns
```

```python
county_and_emissions_df2.groupby(["MSA"]).sum().columns
```

```python
msa_and_emissions_df = county_and_emissions_df2.groupby(["MSA"]).sum()
del(msa_and_emissions_df['FIPS State/County Code'])
del(msa_and_emissions_df['MSA/CMSA Code'])
del(msa_and_emissions_df['FIPS'])
del(msa_and_emissions_df['MSA FIPS'])
msa_and_emissions_df["MSA FIPS"] = county_and_emissions_df2.groupby(["MSA"])["MSA FIPS"].first()
msa_and_emissions_df["Population Density"] = msa_and_emissions_df["Census Population"] /  msa_and_emissions_df["Area (Land)"]
msa_and_emissions_df.sort("Total Emission").head()
```

```python
msa_and_emissions_df[msa_and_emissions_df["MSA FIPS"]==4120] 
```

```python
msa_and_emissions_df["Total Emission per capita"] = 1000000 * msa_and_emissions_df["Total Emission"] / msa_and_emissions_df["Census Population"]
msa_and_emissions_df[["Total Emission","Census Population", "Total Emission per capita"]].sort("Total Emission per capita")
```

```python
plt.rcParams['figure.figsize'] = (10.0, 8.0)
with plt.style.context('fivethirtyeight'):
    plt.scatter(np.log(msa_and_emissions_df["Census Population"]), np.log(1000000*msa_and_emissions_df["Total Emission"]), c="g")
    ax = plt.gca()
    ax.set_title("Total CO2 Emission vs. Population")
    ax.set_xlim(left = 13)
    ax.set_ylim(bottom = 14)    
    ax.set_xlabel("log Population")
    ax.set_ylabel("log Emission in Million Tons")     
    plt.savefig("scatter_log_emission_vs_log_population.jpg")
```

```python
y = np.log(msa_and_emissions_df["Total Emission"]*1000000)
x = np.log(msa_and_emissions_df[["Census Population"]])
x.insert(0, "Offset",1)

model_1 = sm.OLS(y,x).fit()
print model_1.summary()
offset = model_1.params[0]
slope = model_1.params[1]

plt.rcParams['figure.figsize'] = (10.0, 8.0)
with plt.style.context('fivethirtyeight'):
    ax=plt.gca()
    ax.set_aspect('equal')
    ax.set_title("Total CO2 Emission vs. Population")
    ax.set_xlabel("log Population")
    ax.set_ylabel("log Emission in Million Tons")         
    plt.scatter(x['Census Population'], y, c="g")
    line_points =  np.linspace(x['Census Population'].min(), x['Census Population'].max(), 10)
    plt.plot(line_points, slope*line_points + offset)
    plt.savefig("scatter_and_fit_log_emission_vs_log_population.jpg")
```
